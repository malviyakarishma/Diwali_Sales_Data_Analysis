```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as plt
import seaborn as sns

df=pd.read_csv('Diwali Sales Data.csv' , encoding='unicode_escape')
df.head(50)
```

```python
df.isnull()
```

```python
df.info()
```

```python
newdf=df.drop(['Status','unnamed1'],axis=1)
newdf
```

```python
newdf.info()
```

```python
pd.isnull(newdf).sum()
```

```python
newdf.dropna(inplace=True)
```

```python
pd.isnull(newdf).sum()
```

```python
newdf['Amount']=newdf['Amount'].astype('int')
```

```python
newdf['Amount'].dtypes
```

```python
newdf.describe()
```

```python
newdf[['Age','Orders','Amount']].describe()
```

#exploratory_data_analysis

```python
newdf.columns
```

```python
sns.countplot(x='Gender',data=newdf)
```

for adding values on the bars

```python
x_1=sns.countplot(x='Gender',data=newdf)

for bars in x_1.containers:
    x_1.bar_label(bars)
    
```

```python
newdf.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
newdf2=newdf.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=newdf2)
```

```python
x_1=sns.countplot(x='Age Group',hue='Gender',data=newdf)
for bars in x_1.containers:
    x_1.bar_label(bars)
```

```python
newdf2_Age=newdf.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=newdf2_Age)
```

#total amount vs age-group

```python
#total number of orders from top 10 states
newdf_state=newdf.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State',y='Orders',data=newdf_state)
```

```python
#total number of sales/amounts from top 10 states
newdf_state=newdf.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State',y='Amount',data=newdf_state)
```

```python
#maritial status
```

```python
x_1=sns.countplot(x='Marital_Status',data=newdf)
sns.set(rc={'figure.figsize':(2,5)})
for bars in x_1.containers:
    x_1.bar_label(bars)
    
```

```python
newdf_state=newdf.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=newdf_state)
```

```python
x_1=sns.countplot(x='Occupation',data=newdf)
sns.set(rc={'figure.figsize':(30,5)})
for bars in x_1.containers:
    x_1.bar_label(bars)
```

```python
newdf_occu=newdf.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x='Occupation',y='Amount',data=newdf_occu)
```

```python
#total no. of sales from products

x_1=sns.countplot(x='Product_Category',data=newdf)
sns.set(rc={'figure.figsize':(30,5)})
for bars in x_1.containers:
    x_1.bar_label(bars)
```

```python
#purchasing power vs products_category

newdf_pro=newdf.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x='Product_Category',y='Amount',data=newdf_pro)
```
